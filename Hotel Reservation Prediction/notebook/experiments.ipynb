```python
cat_columns = ["type_of_meal_plan", "room_type_reserved", "required_car_parking_space", "room_type_reserved", "market_segment_type", "booking_status"]

num_columns = ["no_of_adults", "no_of_children", "no_of_weekend_nights", "no_of_week_nights", 'lead_time', 'arrival_year', 'arrival_month', 'arrival_date', 'no_of_previous_cancellations', 'no_of_previous_bookings_not_canceled',
       'avg_price_per_room', 'no_of_special_requests']
```

### Univariate Analysis

```python
def num_plot_dist(df, num_featues):
    fig, axes = plt.subplots(len(num_featues),2, figsize=(15, len(num_featues)*5))
    if len(num_featues) == 1:
        axes = [axes]

    for i, column in enumerate(num_featues):
        sns.histplot(data=df, x=column, ax=axes[i][0], kde=True, palette="Blues")
        axes[i][0].set_title(f"Histogram for {column}")
        sns.boxplot(data=df, x=column, ax=axes[i][1], palette="Blues")
        axes[i][1].set_title(f"Box Plot for {column}")

    plt.tight_layout()
    plt.show()

```

```python
num_plot_dist(data, num_columns)
```

```python
for cat_fetaure in cat_columns:
    plt.figure(figsize=(10,6))
    data[cat_fetaure].value_counts().plot(kind='bar', color="skyblue")
    plt.title(f"{cat_fetaure} Distribution")
    plt.xlabel(cat_fetaure)
    plt.ylabel("Count")
    plt.show()
```

### Bivariate Analysis

```python
def plot_bivariate_num(df, target, num_features):
    num_plots = len(num_features)
    num_rows = (num_plots + 1) // 2
    fig, axes = plt.subplots(num_rows, 2, figsize=(15, num_rows*5))
    axes = axes.flatten()
    for i, column in enumerate(num_features):
        sns.boxplot(x=target, y=column, ax=axes[i], data=df, palette="Blues")
        axes[i].set_title(f"{column} vs {target}")

    plt.tight_layout()
    plt.show()
```

```python
plot_bivariate_num(data, 'booking_status', num_columns)
```

```python
sns.barplot(x="arrival_month", y="avg_price_per_room", data=data)
plt.show()
```

```python
def plot_bivariate_cat(df, target, cat_fetaure):
    num_features = len(cat_fetaure)
    num_rows = (num_features + 1)//2
    fig,axes = plt.subplots(num_rows, 2, figsize=(15, num_rows*5))
    axes = axes.flatten()

    for i, feature in enumerate(cat_fetaure):
        sns.countplot(x=feature, hue=target, data=df, palette="Set2", ax=axes[i])
        axes[i].set_title(f"{feature} VS {target}")
        axes[i].tick_params(axis='x', rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
plot_bivariate_cat(data, 'booking_status', cat_columns)
```
